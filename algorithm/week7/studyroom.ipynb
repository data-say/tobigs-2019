```python
import numpy as np
import pandas as pd
from tqdm import trange 
from tqdm import tqdm
import time
import os
```

```python
df = pd.read_csv('study_room_HMS.csv')
df.head()
```

```python
df.shape
```

```python
del df['Unnamed: 0']
```

```python
def split_generator(column):
    i = 0
    for row in column:
        people = row.split(',')
        for person in people:
            yield [i, person]
        i += 1
```

```python
person = split_generator(df['hashed'])
```

```python
# 사람 별 데이터 만들기
lst = []
for i in tqdm(range(100000000)):
    try:
        lst.append(next(person))
    except:
        break
```

```python
data = pd.DataFrame(lst)
data.head()
```
