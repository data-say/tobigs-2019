### a) 1~10의 제곱수로 이루어진 1d array

```python
power = [i ** 2 for i in range(1, 11)]
print(power)
```

### b) string type의 구구단 2d array

```python
gugudan = [['%dx%d=%d' %(i,j,i*j) for j in range(1,10)] for i in range(1,10)]
for row in gugudan:
    print(row)
```

### c) 자기 자신을 곱하는 것을 제외한 string type의 구구단 2d array

```python
gugudan2 = [['%dx%d=%d' %(i,j,i*j) for j in range(1,10) if i != j] for i in range(1,10)]
for row in gugudan2:
    print(row)
```
